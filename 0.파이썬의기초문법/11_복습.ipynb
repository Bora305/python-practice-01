```python
import time, random
c = 0
accel = False
breakpedal = False


while True:
    if c<= 80:
        accel = True
        breakpedal = False
        
    if c>80:
        accel = False
        breakpedal= True
        
    if accel:
        c = c + 7 + random.randrange(0,6)
        
    if breakpedal:
        c = c - 10
        

    
    print(c)
    time.sleep(1)
    
# random 을 가지고 속도의 변화를 주고 싶음
# 7+0~6 이 될 수 있음
```

```python
import time, random

def applecar(c):
    accel = False
    breakpedal = False
    distance = 0
    hour = 0
    
    while True:
        if c<= 80:
            accel = True
            breakpedal = False
            
        if c>80:
            accel = False
            breakpedal= True
            
        if accel:
            c = c + 7 + random.randrange(0,6)
            
        if breakpedal:
            c = c - 10
            
        print(c)
        distance = distance + c
        hour = hour + 1
        
        if distance > 500:
            break
        
        time.sleep(1)
        
    return hour,distance

hour,distance = applecar(70)
print(hour,distance)
```
