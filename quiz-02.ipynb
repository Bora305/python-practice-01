## 문제 1

아래 코드는 데이터를 차트 이미지로 만들어주는 코드이다.

### 문제)

차트에 새로운 장수 `조조`와 새로운 스텟 `통솔`을 추가해 보자.
수치는 임의로 정해서 넣으면 된다.

````python
from IPython.display import display, HTML

def labelRefinery(labels):
  quotedLabels = []
  for label in labels:
    quotedLabels.append("'{}'".format(label));
  return '[{}]'.format(','.join(quotedLabels))

def dataSetRefinery(dataSets):
  refinedDataSets = []
  for dataSet in dataSets:
    refinedDataSets.append("{label:'" + dataSet[0] + "',data:[" + ','.join(map(str, dataSet[1])) + "]}")
  return ','.join(refinedDataSets)

def generateChartUrl(labels, dataSets):
  return "https://quickchart.io/chart?c={type:'bar',data:{labels:" + labelRefinery(labels) + ",datasets:[" + dataSetRefinery(dataSets) + "]}}"

def showImage(url):
  display(HTML('<img src="' + url + '">'))

def drawChart(labels, dataSets):
  url = generateChartUrl(labels, dataSets)
  showImage(url)

def main():
  labels = ['유비', '관우', '장비', '제갈량']
  dataSets = [
    ['무력', [60, 95, 96, 40]],
    ['지력', [70, 65, 50, 100]],
    ['정치', [70, 60, 40, 95]],
    ['매력', [100, 90, 60, 95]],
  ]
  drawChart(labels, dataSets)

main()
````

## 문제 2

아래 코드는 40x15 화면에서 점이 대각선으로 이동하며 경로를 표시하는 코드이다.

### 문제)

각 함수별로 하는 일을 짐작해 보시오. (할 수 있는만큼 디테일 하게)

````python
import time, random, copy
from IPython.display import clear_output

# Initial values
screenSizeX = 40
screenSizeY = 15
screen = [[' '] * screenSizeX for _ in range(screenSizeY)]
x, y, directionX, directionY = 0, 4, 'down', 'left'

def move(x, y, initialDirectionX, initialDirectionY):
  directionX = initialDirectionX
  directionY = initialDirectionY
  if (initialDirectionX == 'up'):
    if x == 0:
      x += 1
      directionX = 'down'
    else:
      x -= 1
  if (initialDirectionX == 'down'):
    if x == screenSizeX - 1:
      x -= 1
      directionX = 'up'
    else:
      x += 1
  if (initialDirectionY == 'left'):
    if y == 0:
      y += 1
      directionY = 'right'
    else:
      y -= 1
  if (initialDirectionY == 'right'):
    if y == screenSizeY - 1:
      y -= 1
      directionY = 'left'
    else:
      y += 1

  return x, y, directionX, directionY

def render(object):
  screen[object[1]][object[0]] = '*'
  clear_output(wait=True)
  print('-' * (screenSizeX + 4))
  for row in screen:
    print('|', ''.join(row), '|')
  print('-' * (screenSizeX + 4))
  lastFootPrint = object
  screen[lastFootPrint[1]][lastFootPrint[0]] = 'O'


def main():
  for i in range(100):
    render([x, y])
    x, y, directionX, directionY = move(x, y, directionX, directionY)
    time.sleep(0.3)


main()
````

## 문제 3

아래 코드는 윷 4개를 던지는 상황을 시뮬레이션 하는 코드이다.

`drawYutSticks` 함수는 던져진 윷을 랜덤으로 생성한다.
- 입력값
  - num: 던질 윷의 개수, 기본값 `4`
- 출력값
  - sticks: 던져진 윷의 결과, 문자열 배열, ex) `['[    ]', '[XXXX]', '[    ]', '[XXXX]']`

`printYutSticks` 함수는 생성된 윷 배열을 화면에 출력한다.

### 문제)

`drawYutSticks` 함수의 출력값에 윷의 족보값 (도, 백도, 개, 걸, 윷, 모)을 추가하고
`printYutSticks` 함수의 입력값으로 전달하여 화면에 출력해보시오.

### 출력 예)

```text
Draw Yut Sticks Game
([ x  ] is for back-do)
----------------------------------
20 | [XXXX] [XXXX] [    ] [XXXX] |
```

````python
import time, random

# 윷 던지기
def drawYutSticks(num = 4):
  sticks = []
  specialStick = random.randint(0, 3);
  for i in range(0, num):
    result = random.choice([' ', 'X'])
    stick = '[XXXX]' if result == ' ' else '{}'.format('[    ]' if i != specialStick else '[ x  ]')
    sticks.append(stick)
  return sticks

# 윷 던지기 결과를 출력합니다.
def printYutSticks(sticks, turn = 0):
  print('{}'.format(turn).rjust(2) if turn != 0 else '', '|', ' '.join(sticks), '|', end = '\r')

print('Draw Yut Sticks Game')
print('([ x  ] is for back-do)')
print('----------------------------------')


def main():
  for x in range (0, 20):
    yutResult = drawYutSticks()
    printYutSticks(yutResult, x + 1)
    time.sleep(1)

main()
````
